```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam,SGD
```

```python
# Load Data
data = pd.read_csv("/content/drive/MyDrive/ionosphere_data.txt")
data.head(4)
```

```python
#Encoding
data['label']=data['label'].map({'g':1,'b':0})
data.head(4)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
X=data.loc[:,data.columns!='label']
Y=data.loc[:,data.columns=='label']
```

```python
print(X.shape,Y.shape)
```

```python
# Normalizing the data
mean = X.mean(axis=0)
X -= mean

std = X.std(axis=0)
X /= std
```

```python
X.head()
```

```python
X.describe()
```

```python
X.feature2.isnull().sum()
```

```python
X1=X.drop(columns='feature2')
```

```python
X1.head(5)
```

```python
X1.shape
```

```python
# Split Data
from sklearn.model_selection import train_test_split
X1_train,X1_test,Y_train,Y_test= train_test_split(X1,Y,test_size=0.40,random_state=8)
```

```python
model = Sequential()
model.add(tf.keras.layers.Dense(16, activation='relu',input_shape=(X1_train.shape[1],)))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['acc'])
```

```python
model.fit(X1_train,Y_train,epochs=100, batch_size=16, verbose=1)
```

```python
prediction=model.predict(X1_test)
M=prediction.round()
M
```
